# Integer Programming

- All decision variables are integers

  $$
  \begin{aligned}
  \text{maximize} \quad & c^T x \\
  \text{subject to} \quad & Ax \le b \\
  & x \ge 0 \\
  & x \in \mathbb{Z}^n
  \end{aligned}
  $$

- Binary integer programming: Variables are restricted to be either 0 or 1

# Binary Knapsack Problem

- Combinatorial optimization problem

- Problem of packing the most valuable or useful items without overloading the luggage.
  - A set of items ($N$ items), each with a weight $w_i$ and a value $v_i$
  - Fixed capacity
  - Maximize the total value possible

## Problem formulation

$$
\begin{aligned}
\text{maximize} \quad & \sum_{i=0}^{N-1} v_i x_i \\
\text{subject to} \quad & \sum_{i=0}^{N-1} w_i x_i \le C \\
& x_i \in \{0,1\} \quad \forall i = 0,\ldots,N-1
\end{aligned}
$$

## Coding in Python

### Creating the data (weights and values)

```python
w = [4,2,5,4,5,1,3,5]
v = [10,5,18,12,15,1,2,8]
c = 15
N = len(w)
```

### Step 2: Importing docplex package

```python
from docplex.mp.model import Model
```

### Step 3: create an optimization model

```python
knapsack_model = Model(name='knapsack')
```

### Step 4: Add multiple binary decision variables

```python
X = knapsack_model.binary_var_list(N, name='x')
```

### Step 5: Add the constraints

```python
knapsack_model.add_constraint(sum(w[i] * X[i] for i in range(N)) <= c, 'capacity')
```

### Step 6: Define the objective function

```python
obj_fun = sum(v[i] * X[i] for i in range(N))
knapsack_model.set_objective('max', obj_fun)

knapsack_model.print_information()
```

### Step 7: Solve the model and output the solution

```python
knapsack_model.solve()
print('Optimization is done. Objective function value: {}'.format(knapsack_model.objective_value))

knapsack_model.print_solution()
```
