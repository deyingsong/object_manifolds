# Quadratically constrained quadratic programming


- Mathematical optimization problems with quadratic constraints and quadratic objective function
- NP-hard problem.
- Can be seen in
  - Optimization of financial portfolios,
  - Image and signal processing,
  - Regression,
  - Scheduling in chemical plants, etc.





## Problem Formulation

Our objective is to find $x \in \mathbb{R}^n$ in the following problem:

$$
\begin{aligned}
\text{minimize} \quad & x^T Q x + q^T x + c, \\
\text{subject to} \quad & x^T Q_i x + q_i^T x + c_i \le b_i, \quad i = 1, \ldots, m
\end{aligned}
$$

where

- $c, c_i \in \mathbb{R}$,
- $Q, Q_i \in \mathbb{R}^{n \times n}$,
- $q, q_i \in \mathbb{R}^n$,
- $b \in \mathbb{R}^m$.

## Coding in Python

The model:

$$
\begin{aligned}
\text{minimize} \quad & x^2 + y^2 + z^2 \\
\text{subject to} \quad 
& x^2 \ge 1, \\
& y^2 \ge 2, \\
& x^2 + y^2 \le 5, \\
& x, y \ge 0, \\
& y \le z, \\
& z \in [0,1].
\end{aligned}
$$

### Step 1: Import package

```python
from docplex.mp.model import Model
```

### Step 2: Create a model

```python
quadratic_model = Model(name='Quadratic')
```

### Step 3: Define decision variables

```python
x = quadratic_model.continuous_var(name='x', lb=0)
y = quadratic_model.integer_var(name='y', lb=0)
z = quadratic_model.binary_var(name='z')
```

### Step 4: Add constraints

```python
# x^2 >= 1
quadratic_model.add_constraint(x*x >= 1, 'x_square')
# y^2 >= 2
quadratic_model.add_constraint(y*y >= 2, 'y_square')
# z^2 + y^2 <= 5
quadratic_model.add_constraint(z*z + y*y <= 5, 'z_y_square')
```

### Step 5: Define the objective function

```python
obj_fn = x*x + y*y + z*z
quadratic_model.set_objective('min', obj_fn)
```

### Step 6: Solve model and output the results

```python
quadratic_model.print_information()
quadratic_model.solve()
```

### Show the solution

```python
quadratic_model.print_solution()
```

### Extras: Add a quadratic constraint

add a quadratic constraint: $x^2 \geq y^2  + z^2$

```python
quadratic_model.add(z**2 + y**2 <= x**2)
quadratic_model.solve()
```
